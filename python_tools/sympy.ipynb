```python
import sympy as sp
```

```python
# Creating a symbol
x = sp.symbols('x')

f = x ** 2 + 4 * x - 16
# Solve this for x=0
sp.solve(f, x)
```

```python
# Solving with equality for a different value
sp.solve(sp.Eq(f, 5), x)
```

```python
sp.plot(f - 5)
```

```python
# Differentiation

sp.diff(f, x)
```

```python
# The derivative is 0 at:

sp.solve(sp.diff(f, x), x)
```

```python
# The function value at -2:

f.subs(x, -2)
```

```python
sp.plot(f)
```

```python
# 2nd derivative of f

sp.diff(sp.diff(f, x), x)
```

```python
# Example

# What is the largest area of a rectangle with sides l and w, where the circumference is fixed

l, w = sp.symbols("l, w")
area = l * w
circumference = 2 * l + 2 * w
```

```python
# Solving the circumference equation to write l as a function of w

print(sp.solve(sp.Eq(circumference, 10), l))
```

```python
# Solving the area equation to w

[w_value] = sp.solve(sp.diff(area.subs(l, -w + 5), w), w)
print(w_value)
```

```python
[l_eq] = sp.solve(sp.Eq(circumference, 10), l)
print(l_eq)
```

```python
l_value = l_eq.subs(w, w_value)
print(l_value)
```

```python
area_value = area.subs({l:l_value, w:w_value})
print(area_value)
```

## Generalization 

```python
# What is the largest area of a rectangle with sides l and w, where the circumference is fixed

l, w = sp.symbols("l, w")
area = l * w
circumference = 2 * l + 2 * w

# parameter for circumference 
c = sp.symbols("c")
```

```python
l_expr = sp.solve(sp.Eq(circumference, c), l)[0]

opt_w = sp.solve(sp.diff(area.subs(l, l_expr), w), w)[0]
opt_l = l_expr.subs(w, opt_w)
opt_area = opt_w * opt_l
print(opt_w, opt_l)
sp.plot(opt_area);
```

## Integration

```python
x = sp.symbols("x")
fx = x + x ** 3 - sp.sin(x)
sp.integrate(fx, x)
```

```python
# Area under the curve

sp.integrate(fx, (x, 1, 3))
```

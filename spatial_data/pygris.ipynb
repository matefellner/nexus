```python
from pygris import tracts
```

```python
tc_counties = ["Hennepin", "Ramsey", "Scott", 
               "Carver", "Dakota", "Washington", "Anoka"]

tc_tracts = tracts(state = "MN", county = tc_counties, 
                   year = 2021, cache = True)

tc_tracts.plot()
```

```python
import numpy as np
import matplotlib.pylab as plt
import gif
```

```python
@gif.frame
def frame(i):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.scatter([x[i]], [y[i]]);

frames = [frame(i) for i in range(100)]
```

```python
gif.save(frames, "sine-wave.gif", duration=100)
```

```python
from sklearn.ensemble import RandomForestRegressor

@gif.frame
def frame(i):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x)
    mod = RandomForestRegressor(n_estimators=1, max_depth=i)
    mod.fit(x.reshape(-1, 1), y)
    plt.plot(x, y)
    plt.plot(x, mod.predict(x.reshape(-1, 1)))
    plt.title(f"max_depth={i}")

frames = [frame(i) for i in range(1,100)]
gif.save(frames, "rf.gif", duration=300)
Image("rf.gif")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def subplots_3d():

    # prepare some coordinates 3d
    x, y, z = np.indices((7, 7, 7))

    # draw cuboids in the top left and bottom right corners
    # and a link between them
    structure_1 = (x < 5) & (y < 5) & (z < 5) & (x > 1) & (y > 1) & (z > 1)

    # combine the objects into a single boolean array
    voxels = structure_1

    # set the colors of each object
    colors = np.empty(voxels.shape, dtype=object)
    colors[structure_1] = "blue"

    # and plot everything

    fig = plt.figure(figsize=plt.figaspect(0.3))

    ax = fig.add_subplot(1, 3, 1, projection="3d")
    ax.voxels(voxels, facecolors=colors, edgecolor="k")
    ax.set_title("node neighbors", x=0.5, y=1)

    ax = fig.add_subplot(1, 3, 2, projection="3d")
    ax.voxels(voxels, facecolors=colors, edgecolor="k")
    ax.set_title("node neighbors", x=0.5, y=1)

    ax = fig.add_subplot(1, 3, 3, projection="3d")
    ax.voxels(voxels, facecolors=colors, edgecolor="k")
    ax.set_title("node neighbors", x=0.5, y=1)

    plt.show()
```

```python
subplots_3d()
```

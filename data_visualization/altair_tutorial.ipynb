```python
import pathlib
import pandas as pd
import altair as alt

df = pd.read_csv("./data/birthdays.csv")

def clean_dataset(dataf):
    return (dataf
            .assign(date = lambda d: pd.to_datetime(d['date']))
            .assign(yday = lambda d: d['date'].dt.dayofyear)
            .groupby(['date', 'wday', 'yday'])
            .agg(births = ('births', 'sum'), month=('month', 'first'))
            .reset_index()
            .head(1000))

plot_df = df.pipe(clean_dataset)
```

```python
plot_df
```

```python
days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']

(alt.Chart(plot_df)
 .mark_point()
 .encode(x='yday', y='births', color=alt.Color('wday', sort=days), tooltip=['date', 'births', 'wday'])
 .properties(width=600, height=300)
 .interactive())
```
